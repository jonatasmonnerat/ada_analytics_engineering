10.  Utilizando uma query em SQL e o banco `sqlite-sakila.db` retorne as cidades e os países, mas apenas os países que começem com a letra **B**.

```python
from sqlalchemy import create_engine
from sqlalchemy import inspect
```

```python
engine = create_engine('sqlite:///./sqlite-sakila.db')
```

```python
insp = inspect(engine)
insp.get_table_names()
```

```python
sql_query ='''
select distinct
    b.city,
    a.country
from 
    country as a
left join
    city as b
on a.country_id = b.country_id
where 
    a.country like "B%"
order by b.city, a.country
    
 '''
df_country_B = pd.read_sql_query(sql_query, engine)
df_country_B
```
